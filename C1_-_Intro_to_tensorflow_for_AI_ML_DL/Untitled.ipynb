```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import seaborn as sb

print(tf.__version__)
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_image, train_label), (test_image, test_label) = fashion_mnist.load_data()
```

```python
xs = np.array([0,1,2,3], dtype=float)
ys = np.array([0.50,1.00,1.50,2.00], dtype=float)

def hist(xs, ys, epoch, pred=None):
    model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(loss='mean_squared_error', optimizer='sgd')
    model.fit(xs, ys, epochs=epoch)
    if pred is not None:
        return model.predict([pred])
    else:
        pass
    
hist(xs, ys, epoch=300, pred=4)
```

```python
xs = np.array([0,2,4,6,8], dtype=float)
ys = np.array([0,4,7,13,16], dtype=float)

hist(xs, ys, epoch=100, pred=10)
```
